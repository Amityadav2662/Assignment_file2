```python
# Q3.What errors are defined in the ArithmeticError class? Explain any two with an example.
Ans.
The ArithmeticError class in Python is the base class for exceptions that occur during arithmetic operations. 
It serves as a superclass for various arithmetic-related exception classes.

The two arithmetic class error is:
1.ZeroDivisionError
2.OverflowError
```

```python
# ZeroDivisionError
a = 10
b = 0

try:
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

```python
# OverflowError
import sys

a = sys.maxsize
b = sys.maxsize

try:
    result = a * b
    print("Result:", result)
except OverflowError:
    print("Error: Overflow occurred during the arithmetic operation.")
```

```python
# Q4.Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans.
The LookupError class in Python is the base class for exceptions that occur when a lookup or indexing operation fails.

KeyError:This exception is raised when a dictionary or dictionary-like object is accessed with a key that does not exist.
IndexError:This exception is raised when a sequence (like a list or a string) is accessed with an invalid index or slice.
```

```python
# KeyError
d = {'name': 'amit', 'age': 20, 'city': 'Delhi'}

try:
    print(d['gender'])
except KeyError:
    print("Error: 'gender' key does not exist.")
```

```python
# IndexError
list = [1, 2, 3, 4, 5]

try:
    print(list[10])
except IndexError:
    print("Error: Index out of range.")
```
